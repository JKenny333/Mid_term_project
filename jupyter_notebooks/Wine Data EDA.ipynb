### Correlations

```python
#Red_wine
corr_matrix_red = red_df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(corr_matrix_red, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.05, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap')  
plt.show()
```

```python
#white
corr_matrix_white = red_df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(corr_matrix_white, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.05, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap')  
plt.show()
```
