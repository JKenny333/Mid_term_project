### Correlations

```python
#Red_wine
corr_matrix_red = red_df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(corr_matrix_red, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.05, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap')  
plt.show()
```

```python
#white
corr_matrix_white = red_df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(corr_matrix_white, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.05, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap')  
plt.show()
```

```python
# Question: To what extent do the correlation heatmaps differ for red and white wine? 

import numpy as np # Comparison of data sets (red-white)

frobenius_norm = np.linalg.norm(corr_matrix_red - corr_matrix_white, 'fro')
print(f"Frobenius Norm: {frobenius_norm}")
```

```python
frobenius_norm = np.linalg.norm(corr_matrix_white - corr_matrix_red, 'fro')
print(f"Frobenius Norm: {frobenius_norm}")
```

```python
# A Frobenius norm of 0.0 indicates that the matrices being compared are identical, 
# or at least, very close to being identical in terms of their element-wise values. 
# The Frobenius norm is a measure of the magnitude of a matrix and is calculated as 
# the square root of the sum of the squares of its elements.

# On this basis, we assume that we can concatenate the datasets for the purpose of our analysis 
# with limited to no impact on our model.
```
