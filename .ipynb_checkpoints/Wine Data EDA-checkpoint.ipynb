```python
url_red = 'https://raw.githubusercontent.com/JKenny333/Mid_term_project/main/csv_files/winequality-red%20-%20winequality-red.csv.csv'
url_white = 'https://raw.githubusercontent.com/JKenny333/Mid_term_project/main/csv_files/winequality-white%20-%20winequality-white.csv.csv'
red_df = pd.read_csv(url_red)
white_df = pd.read_csv(url_white)
red_df
white_df
```

### Correlations

```python
corr_matrix_red = red_df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(corr_matrix_red, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.05, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap')  
plt.show()
```

```python
corr_matrix_white = red_df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(corr_matrix_white, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.05, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap')  
plt.show()
```
